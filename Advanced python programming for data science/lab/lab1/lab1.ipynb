Using a constructor to create an object

```python
#default constructor
class student:
    def __init__(self):
        print("Hello World")
std = student()
```

```python
#parameterized constructor
class student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
std = student("Alice", 22)
std.display()
```

Single inhertiance

```python
# parent class
class animal:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} makes a sound")

# child class
class Dog(animal):
    def speak(self):
        print(f"{self.name} barks")

#creating objects
a = animal("Generic Animal")
d = Dog("buddy")

a.speak() #generic animal makes a sound
d.speak() #buddy barks
 
```

```python
class animal:
    def __init__(self,name):
        self.name = name
    def display(self):
        print(self.name)
class sound:
    def __init__(self,speak):
        self.speak = speak
    def display(self):
        print(self.speak)
class dog(animal,sound):
    def __init__(self,name,speak):
        animal.__init__(self,name)
        sound.__init__(self,speak)
    def display(self):
        print(self.name)
        print(self.speak)
d = dog("Buddy","Bark")
d.display()
```

Multiple inheritance

```python
class CanRun:
    def run(self):
        print("i can run fast")

class CanSwim:
    def swim(self):
        print("i can swim well")

class Dog(CanRun,CanSwim):
    pass

d=Dog()
d.run() #i can run fast
d.swim() #i can swim well
```

Using super()

```python
class Dog(animal):
    def __init__(self,name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        print(f"{self.name},the {self.breed}, barks")

d = Dog("Buddy", "Labrador")
d.speak()
```
