4. Create a custom iterator that generates squares of numbers up to a given limit.

```python
# Custom iterator to generate squares of numbers up to a given limit

class Squares:
    def __init__(self, limit):
        self.limit = limit
        self.num = 1   # start from 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= self.limit:
            result = self.num ** 2
            self.num += 1
            return result
        else:
            raise StopIteration

# Example usage
limit = 5
squares_iterator = Squares(limit)

for square in squares_iterator:
    print(square)
```

# C. Generator

## 3. Generators

        Generators are special functions or objects used to produce values one at a time instead of storing all values in memory at once. They are mainly used for efficient iteration over large data sets.

```python
#Generator function
def my_generator():
    yield 1
    yield 2
#Using generator
for value in my_generator():
    pass
#Generator expression
gen = (x for x in range(5))
```

1. Write a generator function to generate numbers from 1 to n.

```python
def generate_numbers(n):
    for i in range(1, n + 1):
        yield i
for num in generate_numbers(5):
    print(num)
```

2. Write a generator that yields squares of numbers up to a given limit.

```python
def generate_squares(limit):
    for i in range(1, limit + 1):
        yield i ** 2

for square in generate_squares(5):
    print(square)
```

3. Write a generator to generate Fibonacci numbers up to n.

```python
def generate_fibonacci(n):
    a, b = 0, 1
    while a <= n:
        yield a
        a, b = b, a + b
print("Fibonacci numbers up to 50:")
for num in generate_fibonacci(50):
    print(num)
```

4. Write a generator that yields only prime numbers within a given range.

```python
def generate_primes(start, end):
    def is_prime(num):
        if num < 2:
            return False
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                return False
        return True

    for n in range(start, end + 1):
        if is_prime(n):
            yield n

print("Prime numbers between 10 and 50:")
for prime in generate_primes(10, 50):
    print(prime)
```

# D. Decorators

## 4. Decorators

        Decorators are a design pattern / programming feature used to add new functionality to an existing object or function without modifying its structure.


```python
#Basic decorator
def my_decorator(func):
    def wrapper():
        pass
    return wrapper
@my_decorator
def my_function():
    pass
```

1. Write a decorator that prints a message before and after a function execution.

```python
def message_decorator(func):
    def wrapper(*args, **kwargs):
        print(">>> Before function execution")
        result = func(*args, **kwargs)
        print(">>> After function execution")
        return result
    return wrapper

@message_decorator
def greet(name):
    print(f"Hello, {name}!")

# Calling the decorated function
greet("John")
```

2. Write a decorator that counts how many times a function is called.

```python
def count_calls(func):
    def wrapper(*args, **kwargs):
        wrapper.calls += 1
        print(f"Function '{func.__name__}' has been called {wrapper.calls} times.")
        return func(*args, **kwargs)
    wrapper.calls = 0
    return wrapper

@count_calls
def greet(name):
    print(f"Hello, {name}!")

greet("john")
greet("doe")
greet("sam")
```

3. Write a decorator that converts the return value of a function to uppercase.

```python
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        if isinstance(result, str):   # Only convert if return value is a string
            return result.upper()
        return result
    return wrapper

@uppercase_decorator
def greet(name):
    return f"Hello, {name}"

print(greet("Ram"))
```
