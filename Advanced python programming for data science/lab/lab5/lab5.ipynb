2. Generate two random numerical arrays and compute:
- Covariance
- Pearson correlation coefficient
- Spearman correlation coefficient

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Step 1: Generate two random numerical arrays
np.random.seed(42)  # For reproducibility
x = np.random.rand(100)
y = np.random.rand(100)

# Step 2: Covariance
covariance_matrix = np.cov(x, y)
covariance = covariance_matrix[0, 1]

# Step 3: Pearson Correlation
pearson_corr, _ = pearsonr(x, y)

# Step 4: Spearman Correlation
spearman_corr, _ = spearmanr(x, y)

# Step 5: Display results
print("Array X:\n", x)
print("\nArray Y:\n", y)

print("\nCovariance:", covariance)
print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman Correlation Coefficient:", spearman_corr)
```

8. From a dataset of 1000 random values:
- Take a random sample of 100
- Take a systematic sample
- Take a stratified sample

```python
import numpy as np
import pandas as pd

# Step 1: Generate dataset of 1000 random values
np.random.seed(42)
data = np.random.randint(1, 101, 1000)

# Convert to DataFrame for stratified sampling
df = pd.DataFrame({'Values': data})

# -------------------------------------------------
# 1️⃣ Random Sampling (Simple Random Sample)
random_sample = np.random.choice(data, size=100, replace=False)

# -------------------------------------------------
# 2️⃣ Systematic Sampling
k = len(data) // 100   # Sampling interval
systematic_sample = data[::k][:100]

# -------------------------------------------------
# 3️⃣ Stratified Sampling
# Create strata (divide values into 4 groups based on value ranges)
df['Strata'] = pd.cut(df['Values'], bins=4, labels=False)

# Take proportional samples from each stratum
stratified_sample = df.groupby('Strata', group_keys=False).apply(
    lambda x: x.sample(frac=0.1)
)

# -------------------------------------------------
# Print Results
print("Random Sample (first 10 values):")
print(random_sample[:10])

print("\nSystematic Sample (first 10 values):")
print(systematic_sample[:10])

print("\nStratified Sample (first 10 values):")
print(stratified_sample.head(10))
```

9. Demonstrate the Central Limit Theorem by:
- Taking 50 different samples
- Plotting the sampling distribution of the mean

```python
# Step 1: Create a population (non-normal distribution for demonstration)
np.random.seed(42)
population = np.random.exponential(scale=2, size=1000)

# Step 2: Take 50 different samples and calculate their means
sample_size = 30
num_samples = 50
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Step 3: Plot sampling distribution of the mean
plt.figure()
plt.hist(sample_means, bins=10, edgecolor='black')
plt.title("Sampling Distribution of the Mean (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# Step 4: Display mean and standard deviation of sample means
print("Mean of sample means:", np.mean(sample_means))
print("Standard deviation of sample means:", np.std(sample_means))
```

10. Perform a one-sample t-test on a dataset to check if the population mean equals a given
value.

```python
import numpy as np
from scipy.stats import ttest_1samp

# Step 1: Generate dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=30)  # Sample of 30 values

# Step 2: Define the population mean to test
population_mean = 52

# Step 3: Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, population_mean)

# Step 4: Display results
print("Sample Data (first 10 values):", data[:10])
print("\nT-statistic:", t_stat)
print("P-value:", p_value)

# Step 5: Interpret result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: sample mean is significantly different from", population_mean)
else:
    print("Fail to reject the null hypothesis: sample mean is not significantly different from", population_mean)
```

11. Perform an independent two-sample t-test between two datasets.

```python
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Generate two independent datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=30)  # Sample 1
data2 = np.random.normal(loc=52, scale=5, size=30)  # Sample 2

# Step 2: Perform independent two-sample t-test
t_stat, p_value = ttest_ind(data1, data2, equal_var=True)  # assume equal variances

# Step 3: Display results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Step 4: Interpret result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: the means of the two samples are significantly different")
else:
    print("Fail to reject the null hypothesis: no significant difference between the sample means")
```

12. Perform a chi-square test for independence on a categorical dataset.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a sample categorical dataset
# Example: Survey of 50 people by Gender and Preference
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'] * 6 + ['Male', 'Female'],
        'Preference': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A'] * 6 + ['A', 'B']}

df = pd.DataFrame(data)

# Step 2: Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])

print("Contingency Table:")
print(contingency_table)

# Step 3: Perform Chi-Square Test of Independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Step 4: Display results
print("\nChi-square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Step 5: Interpret result
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: Gender and Preference are dependent")
else:
    print("Fail to reject the null hypothesis: Gender and Preference are independent")
```

13. Calculate and interpret a 95% confidence interval for the mean.

```python
import numpy as np
from scipy import stats

# Step 1: Generate a sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=30)  # Sample of 30 values

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)
standard_error = sample_std / np.sqrt(n)

# Step 3: Determine t critical value for 95% confidence
confidence_level = 0.95
alpha = 1 - confidence_level
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 4: Calculate confidence interval
margin_of_error = t_crit * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 5: Display results
print("Sample Mean:", sample_mean)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

14. Create a simple linear regression model using statsmodels.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(42)
X = np.random.rand(50) * 10  # Independent variable
y = 2.5 * X + np.random.normal(0, 5, 50)  # Dependent variable with some noise

# Step 2: Add constant term for intercept
X_with_const = sm.add_constant(X)  # Adds a column of 1s for intercept

# Step 3: Fit linear regression model
model = sm.OLS(y, X_with_const)
results = model.fit()

# Step 4: Display summary of the regression
print(results.summary())

# Step 5: Plot regression line
plt.scatter(X, y, label="Data")
plt.plot(X, results.predict(X_with_const), color='red', label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

15. Interpret regression output: R-squared, p-value, coefficients.

```python
import numpy as np
import statsmodels.api as sm

# Step 1: Generate sample data
np.random.seed(42)
X = np.random.rand(50) * 10  # Independent variable
y = 2.5 * X + np.random.normal(0, 5, 50)  # Dependent variable with noise

# Step 2: Add constant term for intercept
X_with_const = sm.add_constant(X)

# Step 3: Fit linear regression model
model = sm.OLS(y, X_with_const)
results = model.fit()

# Step 4: Extract R-squared, coefficients, and p-values
r_squared = results.rsquared
intercept, slope = results.params
p_values = results.pvalues

# Step 5: Print interpretation
print("=== Regression Output Interpretation ===\n")
print(f"R-squared: {r_squared:.3f}")
print(f"Intercept: {intercept:.3f}, Slope: {slope:.3f}")
print(f"P-values:\nIntercept: {p_values[0]:.3f}, Slope: {p_values[1]:.3f}\n")

# Step 6: Interpretation logic
alpha = 0.05
if p_values[1] < alpha:
    print("The slope is statistically significant: X has a significant effect on y.")
else:
    print("The slope is not statistically significant: X may not affect y.")

print(f"\nInterpretation of R-squared: {r_squared:.3f} of the variance in y is explained by X.")
```

16. Perform multiple linear regression with at least two independent variables.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Step 1: Generate sample data
np.random.seed(42)
n_samples = 100

X1 = np.random.rand(n_samples) * 10       # Independent variable 1 (e.g., feature1)
X2 = np.random.rand(n_samples) * 20       # Independent variable 2 (e.g., feature2)

# Dependent variable (y) generated using a linear combination of X1 and X2 + noise
# Equation: y = 3 + 2*X1 + 0.5*X2 + random_noise
y = 3 + 2*X1 + 0.5*X2 + np.random.normal(0, 5, n_samples)

# Step 2: Create DataFrame
df = pd.DataFrame({'X1': X1, 'X2': X2, 'y': y})

# Step 3: Define independent (X) and dependent (y) variables
X = df[['X1', 'X2']]  # Predictor variables
y = df['y']            # Response variable

# Step 4: Add constant term to include intercept in the regression
X_with_const = sm.add_constant(X)  # Intercept term corresponds to the "3" in the equation

# Step 5: Fit multiple linear regression model
model = sm.OLS(y, X_with_const)  # OLS = Ordinary Least Squares
results = model.fit()

# Step 6: Display regression summary (includes coefficients, R-squared, p-values)
print(results.summary())

# Step 7: Optional: Plot actual vs predicted values
y_pred = results.predict(X_with_const)
plt.scatter(y, y_pred)
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.title("Actual vs Predicted Values")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # 45-degree line
plt.show()
```

17. Generate time-series data and perform trend analysis using linear regression.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 1: Generate time-series data
np.random.seed(42)
n_periods = 50
time = np.arange(n_periods)  # Time variable (0,1,2,...,49)

# Simulate data with an upward trend + noise
# Equation: y = 5 + 0.8*time + random_noise
y = 5 + 0.8*time + np.random.normal(0, 3, n_periods)

# Step 2: Convert to DataFrame
df = pd.DataFrame({'Time': time, 'Value': y})

# Step 3: Fit linear regression for trend analysis
X = sm.add_constant(df['Time'])  # Add intercept
model = sm.OLS(df['Value'], X)
results = model.fit()

# Step 4: Display regression summary
print(results.summary())

# Step 5: Plot time series and trend line
plt.figure(figsize=(10,5))
plt.plot(df['Time'], df['Value'], label='Time Series Data', marker='o')
plt.plot(df['Time'], results.predict(X), color='red', label='Trend Line', linewidth=2)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Trend Analysis')
plt.legend()
plt.show()
```

19. Write a Python function that:
- Loads a dataset
- Displays summary statistics
- Plots histogram for all numerical columns

```python
import pandas as pd
import matplotlib.pyplot as plt

def explore_dataset(file_path):
    """
    Loads a dataset, displays summary statistics,
    and plots histograms for all numerical columns.
    
    Parameters:
    file_path (str): Path to CSV dataset
    """
    # Step 1: Load dataset
    df = pd.read_csv(file_path)
    
    # Step 2: Display first few rows
    print("First 5 rows of the dataset:\n")
    print(df.head(), "\n")
    
    # Step 3: Display summary statistics
    print("Summary statistics:\n")
    print(df.describe(), "\n")
    
    # Step 4: Plot histogram for all numerical columns
    numeric_cols = df.select_dtypes(include='number').columns
    df[numeric_cols].hist(bins=20, figsize=(15, 5), layout=(1, len(numeric_cols)))
    
    plt.suptitle("Histograms of Numerical Columns", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Example usage:
# explore_dataset('your_dataset.csv')
explore_dataset('Chocolate Sales (2).csv')
```
