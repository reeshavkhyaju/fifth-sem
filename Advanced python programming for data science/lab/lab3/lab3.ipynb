4. Write a program to access a SOAP web service using Python.

```python
from zeep import Client

wsdl_url = "http://www.dneonline.com/calculator.asmx?WSDL"

client = Client(wsdl=wsdl_url)

print("Available SOAP operations:")
for service in client.wsdl.services.values():
    for port in service.ports.values():
        operations = port.binding._operations
        for op in operations:
            print(op)
```

5. Write a Python program to perform an arithmetic operation using a SOAP API.

```python
from zeep import Client

# SOAP service WSDL URL
wsdl_url = "http://www.dneonline.com/calculator.asmx?WSDL"

# Create SOAP client
client = Client(wsdl=wsdl_url)

# Input values
a = 12
b = 8

# Perform arithmetic operation (Addition)
result = client.service.Add(intA=a, intB=b)

# Display result
print("Addition Result:", result)
```

6. Write a program to handle HTTP status codes while accessing an API.

```python
import requests

# API URL (sample public API)
url = "https://jsonplaceholder.typicode.com/posts/1"

# Send GET request
response = requests.get(url)

# Handle HTTP status codes
if response.status_code == 200:
    print("Success!")
    print("Response Data:", response.json())

elif response.status_code == 404:
    print("Error 404: Resource not found")

elif response.status_code == 401:
    print("Error 401: Unauthorized access")

elif response.status_code == 500:
    print("Error 500: Server error")

else:
    print("Error:", response.status_code)
```

10. Write a program to scrape tabular data from a website.

```python
table = soup.find("table", class_="wikitable")
table_headers = [th.get_text(strip= True) for th in table.find_all("th")]
print("Headers:", table_headers)

table_rows = []
for tr in table.find_all("tr")[1:]:
    cells = [td.get_text(strip=True) for td in tr.find_all("td")]
    if cells:
        table_rows.append(cells)

for row in table_rows[:5]:
    print(row)
```

14. Write a program to process large datasets without loading them fully into memory.

```python
import pandas as pd
import os

# Input and output files
csv_file = "fifa_player_performance_market_value.csv"      # large dataset
output_file = "high_value_players.csv"
chunk_size = 5000


def process_chunk(chunk):
    # Clean column names
    chunk.columns = chunk.columns.str.strip().str.lower()

    # Required columns for processing
    required_columns = [
        'player_id', 'player_name', 'overall_rating',
        'market_value_million_eur', 'transfer_risk_level',
        'club', 'position'
    ]

    # Check if columns exist
    for col in required_columns:
        if col not in chunk.columns:
            print(f"Warning: Column '{col}' missing in this chunk")
            return pd.DataFrame()

    # Filter conditions
    filtered = chunk[
        (chunk['overall_rating'] >= 85) &
        (chunk['transfer_risk_level'] != 'High')
    ]

    # Select important columns
    selected = filtered[
        ['player_id', 'player_name', 'club', 'position',
         'overall_rating', 'market_value_million_eur']
    ]

    return selected


# Remove old output file if exists
if os.path.exists(output_file):
    os.remove(output_file)

print("Starting large dataset processing...\n")

# Read CSV in chunks
for i, chunk in enumerate(pd.read_csv(csv_file, chunksize=chunk_size), start=1):
    print(f"Processing chunk {i}...")

    result = process_chunk(chunk)

    if not result.empty:
        result.to_csv(
            output_file,
            mode='a',
            header=not os.path.exists(output_file),
            index=False
        )

print("\nFinished processing all chunks!")
print(f"Processed data saved to: {output_file}")
```

15. Write a Python program to calculate total rows using chunk-based reading.

```python
import pandas as pd

# Path to large CSV file
csv_file = "fifa_player_performance_market_value.csv"   # change to your file name

# Define chunk size
chunk_size = 5000

# Initialize counter
total_rows = 0

print("Counting rows using chunk processing...\n")

# Read file in chunks
for i, chunk in enumerate(pd.read_csv(csv_file, chunksize=chunk_size), start=1):
    rows_in_chunk = len(chunk)
    total_rows += rows_in_chunk
    print(f"Chunk {i}: {rows_in_chunk} rows")

# Display total rows
print("\nTotal number of rows in dataset:", total_rows)
```

16. Write a program to demonstrate lazy evaluation using generators.

```python
# Generator function demonstrating lazy evaluation
def generate_numbers(n):
    print("Generator started")
    
    for i in range(1, n + 1):
        print(f"Producing value: {i}")
        yield i   # Value is produced only when requested

# Using the generator
gen = generate_numbers(5)

# Values are generated lazily
for num in gen:
    print(f"Consumed value: {num}")
```

17. Write a Python program to read a large text file using generators.

```python
# Generator function to read file line by line
def read_large_file(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            yield line.strip()   # Yield one line at a time


# File path
file_name = "hyperlinks.txt"  

# Using the generator
for line in read_large_file(file_name):
    print(line)
```

18. Write a program to compare normal file reading and lazy evaluation.

```python
import time

file_name = "hyperlinks.txt"  # Replace with your large text file

# --------- Normal File Reading ---------
start_time = time.time()

with open(file_name, "r", encoding="utf-8") as f:
    lines = f.readlines()  # Loads entire file into memory

# Example processing: count lines
total_lines = len(lines)

end_time = time.time()
print("Normal Reading:")
print("Total lines:", total_lines)
print("Time taken:", round(end_time - start_time, 4), "seconds\n")


# --------- Lazy Evaluation Using Generator ---------
def read_file_lazy(file_path):
    """Generator to read file line by line"""
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            yield line

start_time = time.time()

line_count = 0
for _ in read_file_lazy(file_name):
    line_count += 1

end_time = time.time()
print("Lazy Reading (Generator):")
print("Total lines:", line_count)
print("Time taken:", round(end_time - start_time, 4), "seconds")
```
