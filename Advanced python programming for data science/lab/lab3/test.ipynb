## 1. Accessing and Processing Data from APIs(REST,SOAP)

```python
import requests
from bs4 import BeautifulSoup
url="https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) ' 
        
}

response = requests.get(url,headers=headers)
print(response.status_code)
soup = BeautifulSoup(response.text,"html.parser")

movies = soup.find_all("div",class_='ipc-metadata-list-summary-item__c')
print(movies)

for movie in movies:
    title_tag= movie.find("h3",class_='ipc-title__text').text
    rating_tag=movie.find("span",class_="ipc-rating-star--rating").text
    print(title_tag,rating_tag)
```

1. Write a Python program to consume a REST API and display JSON data.

```python
import requests
import json

def fetch_api_data():
    # Example REST API (JSONPlaceholder provides free fake API data)
    url = "https://jsonplaceholder.typicode.com/posts"

    try:
        # Send GET request
        response = requests.get(url)

        # Check if request was successful
        if response.status_code == 200:
            # Parse JSON data
            data = response.json()

            # Display JSON data in a readable format
            print(json.dumps(data, indent=4))
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error occurred: {e}")

if __name__ == "__main__":
    fetch_api_data()
```

2. Write a program to fetch data from a REST API and count total records.

```python
import requests

def count_api_records():
    # Example REST API (JSONPlaceholder provides free fake API data)
    url = "https://jsonplaceholder.typicode.com/posts"

    try:
        # Send GET request
        response = requests.get(url)

        # Check if request was successful
        if response.status_code == 200:
            # Parse JSON data
            data = response.json()

            # Count total records
            total_records = len(data)

            print(f"Total records fetched: {total_records}")
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error occurred: {e}")

if __name__ == "__main__":
    count_api_records()
```

3. Write a Python program to extract specific fields from REST API response.

```python
import requests

def extract_fields_from_api():
    # Example REST API (JSONPlaceholder provides free fake API data)
    url = "https://jsonplaceholder.typicode.com/posts"

    try:
        # Send GET request
        response = requests.get(url)

        # Check if request was successful
        if response.status_code == 200:
            # Parse JSON data
            data = response.json()

            # Extract specific fields (title and body)
            for record in data[:5]:  # limit to first 5 records for readability
                print(f"ID: {record['id']}")
                print(f"Title: {record['title']}")
                print(f"Body: {record['body']}\n")
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error occurred: {e}")

if __name__ == "__main__":
    extract_fields_from_api()
```

4. Write a program to access a SOAP web service using Python.

```python
from zeep import Client

# WSDL URL of the SOAP service
wsdl_url = "http://www.dneonline.com/calculator.asmx?WSDL"

# Create a SOAP client
client = Client(wsdl=wsdl_url)

# Call SOAP methods
num1 = 10
num2 = 5

add_result = client.service.Add(intA=num1, intB=num2)
sub_result = client.service.Subtract(intA=num1, intB=num2)
mul_result = client.service.Multiply(intA=num1, intB=num2)
div_result = client.service.Divide(intA=num1, intB=num2)

# Display results
print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
```

5. Write a Python program to perform an arithmetic operation using a SOAP API.

```python
from zeep import Client

# SOAP service WSDL URL
wsdl_url = "http://www.dneonline.com/calculator.asmx?WSDL"

# Create SOAP client
client = Client(wsdl=wsdl_url)

# Input values
a = 12
b = 8

# Perform arithmetic operation (Addition)
result = client.service.Add(intA=a, intB=b)

# Display result
print("Addition Result:", result)
```

6. Write a program to handle HTTP status codes while accessing an API.

```python
import requests

# API URL (sample public API)
url = "https://jsonplaceholder.typicode.com/posts/1"

# Send GET request
response = requests.get(url)

# Handle HTTP status codes
if response.status_code == 200:
    print("Success!")
    print("Response Data:", response.json())

elif response.status_code == 404:
    print("Error 404: Resource not found")

elif response.status_code == 401:
    print("Error 401: Unauthorized access")

elif response.status_code == 500:
    print("Error 500: Server error")

else:
    print("Error:", response.status_code)
```

8. Write a program to extract the title of a web page using BeautifulSoup.

```python
import requests
from bs4 import BeautifulSoup

# IMDb Top 250 URL
url = "https://www.imdb.com/chart/top/"

# Send HTTP request
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find movie title elements
movies = soup.select("td.titleColumn a")

print("IMDb Top Movies:\n")

# Extract and print movie titles
for i, movie in enumerate(movies[:10], start=1):  # top 10 only
    print(f"{i}. {movie.text}")
```

9. Write a Python program to extract all hyperlinks from a webpage.

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

for link in soup.find_all("a"):
    print(link.get("href"))
```
