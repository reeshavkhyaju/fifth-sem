1. Pandas Profiling

```python
# Step 1: Import Libraries
import pandas as pd
from ydata_profiling import ProfileReport

# Step 2: Load Dataset
df = pd.read_csv("data/Chocolate Sales (2).csv")

# Step 3: Data Cleaning

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

# Remove $ and commas from Amount and convert to float
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Step 4: Verify Cleaned Data
print(df.info())
print(df.head())

# Step 5: Generate Profiling Report
profile = ProfileReport(
    df,
    title="Chocolate Sales EDA Report",
    explorative=True
)

# Step 6: Save Report as HTML
profile.to_file("Chocolate_Sales_EDA_Report.html")

print("EDA Report Generated Successfully!")
```

```python
# Step 1: Import Libraries
import pandas as pd
import sweetviz as sv

# Step 2: Load Dataset
df = pd.read_csv("data/Chocolate Sales (2).csv")

# Step 3: Data Cleaning

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

# Clean Amount column (remove $ and commas)
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Step 4: Verify Data
print(df.info())
print(df.head())

# Step 5: Generate Sweetviz Report
report = sv.analyze(df)

# Step 6: Save Report as HTML
report.show_html("Chocolate_Sales_Sweetviz_Report.html")

print("Sweetviz EDA Report Generated Successfully!")
```

```python
# Step 1: Import libraries
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class

# Step 2: Load dataset
df = pd.read_csv("data/Chocolate Sales (2).csv")

# Step 3: Clean columns if needed
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df["Amount"] = df["Amount"].replace('[\$,]', '', regex=True).astype(float)

# Step 4: Initialize AutoViz
AV = AutoViz_Class()

# Step 5: Run AutoViz (generates EDA plots)
# filename="" since we're using dataframe directly
df_AV = AV.AutoViz(
    filename="", 
    sep=",", 
    depVar="",  # No dependent variable for pure EDA
    dfte=df, 
    header=0, 
    verbose=1,
    save_plot_dir="autoviz_report"  # directory to save plots
)

# Step 6: Create HTML report
import os

html_file = "Chocolate_Sales_AutoViz_Report.html"
with open(html_file, "w") as f:
    f.write("<html><head><title>Chocolate Sales EDA Report</title></head><body>\n")
    f.write("<h1>Chocolate Sales EDA Report (AutoViz)</h1>\n")
    f.write("<p>All plots generated by AutoViz are below:</p>\n")
    
    # Add all saved plot images
    plot_dir = "autoviz_report"
    if os.path.exists(plot_dir):
        for img_file in sorted(os.listdir(plot_dir)):
            if img_file.endswith(".png"):
                f.write(f'<h3>{img_file}</h3>\n')
                f.write(f'<img src="{plot_dir}/{img_file}" width="800"><br><br>\n')
    
    f.write("</body></html>")

print(f"AutoViz EDA done! HTML report saved as: {html_file}")
```
