Q. Why we need to install nbformat and ipykernel for Plotly in VS Code Jupyter Notebook

When using Plotly in a VS Code Jupyter notebook (.ipynb), you might get this error:

ValueError: Mime type rendering requires nbformat>=4.2.0


To fix this, we run:

pip install nbformat ipykernel

1️⃣ nbformat

Jupyter notebooks are stored as JSON files.

nbformat allows Python to read, write, and render notebook content.

Plotly requires it to display interactive charts inline in the notebook.

Without it, Plotly cannot render inline and tries to open a browser instead.

2️⃣ ipykernel

Connects your Python environment (venv) to Jupyter.

Lets the notebook execute Python code from your current environment.

Required for Plotly and other libraries to run correctly in VS Code notebooks.

| Package     | Purpose                                                   |
| ----------- | --------------------------------------------------------- |
| `nbformat`  | Allows Jupyter to render notebook content (inline charts) |
| `ipykernel` | Connects your Python environment to the notebook          |

# 1. matplotlib.pdf

1. Line plot

```python
import plotly.express as px

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 180, 170, 200, 220]

fig = px.line(
    x=months,
    y=sales,
    markers=True,
    title="Monthly Sales of the Company",
    labels={"x": "Months", "y": "Sales"}
)

fig.show()
```

2. Bar Chart

```python
import plotly.express as px

# Given data
departments = ['CSE', 'ECE', 'ME', 'CE']
students = [120, 90, 60, 50]

# Create bar chart
fig = px.bar(
    x=departments,
    y=students,
    title='Number of Students in Different Departments',
    labels={'x': 'Departments', 'y': 'Number of Students'}
)

# Display the chart (inline in VS Code notebook)
fig.show()
```

3. Histogram

```python
import plotly.express as px

# Given marks
marks = [45, 55, 60, 70, 75, 80, 85, 90, 65, 68, 72, 78]

# Create histogram
fig = px.histogram(
    x=marks,
    nbins=5,  # number of bins
    title="Histogram of Students Marks",
    labels={"x": "Marks", "y": "Number of Students"}
)

# Display the histogram inline in VS Code notebook
fig.show()
```

4. Scatter Plot 

```python
import plotly.express as px

# Given data
hours_studied = [1, 2, 3, 4, 5, 6]
marks = [35, 40, 50, 65, 75, 85]

# Create scatter plot
fig = px.scatter(
    x=hours_studied,
    y=marks,
    title='Relationship Between Hours Studied and Marks Obtained',
    labels={'x': 'Hours Studied', 'y': 'Marks Obtained'}
)

# Display the plot inline in VS Code notebook
fig.show()
```

5. Subplots

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Data for Subplot 1: Monthly Sales (Line Plot)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 180, 170, 200, 220]

# Data for Subplot 2: Department-wise Students (Bar Chart)
departments = ['CSE', 'ECE', 'ME', 'CE']
students = [120, 90, 60, 50]

# Create subplots: 1 row, 2 columns
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Monthly Sales', 'Department-wise Student Strength')
)

# Add line plot to first subplot
fig.add_trace(
    go.Scatter(x=months, y=sales, mode='lines+markers', name='Sales'),
    row=1, col=1
)

# Add bar chart to second subplot
fig.add_trace(
    go.Bar(x=departments, y=students, name='Students'),
    row=1, col=2
)

# Update axes labels
fig.update_xaxes(title_text="Months", row=1, col=1)
fig.update_yaxes(title_text="Sales", row=1, col=1)
fig.update_xaxes(title_text="Departments", row=1, col=2)
fig.update_yaxes(title_text="Number of Students", row=1, col=2)

# Update layout
fig.update_layout(
    title_text="Company Sales & Department Student Strength",
    showlegend=False,
    width=900, height=400
)

# Display the figure
fig.show()
```

# 2. Seaborn.pdf

1. Box Plot

```python
import plotly.express as px

# Given data
marks = {
    'Mathematics': [65, 70, 75, 80, 85, 90],
    'Physics': [60, 68, 72, 78, 82, 88],
    'Chemistry': [62, 67, 74, 79, 84, 89]
}

# Convert data to long form for Plotly
data = []
for subject, scores in marks.items():
    for score in scores:
        data.append({'Subject': subject, 'Marks': score})

# Create box plot
fig = px.box(
    data, 
    x='Subject', 
    y='Marks',
    title='Distribution of Marks in Different Subjects'
)

# Display the plot inline (VS Code Jupyter)
fig.show()
```

2. Pair Plot

```python
import plotly.express as px
import pandas as pd

# Creating the dataset
data = {
    'Height': [150, 155, 160, 165, 170],
    'Weight': [50, 55, 60, 65, 70],
    'Age': [18, 19, 20, 21, 22]
}

df = pd.DataFrame(data)

# Creating the scatter matrix (pair plot)
fig = px.scatter_matrix(
    df,
    dimensions=['Height', 'Weight', 'Age'],
    title='Pair Plot of Height, Weight, and Age'
)

# Display the plot inline in VS Code Jupyter
fig.show()
```

3. Heat Map

```python
import plotly.express as px
import pandas as pd

# Sample dataset
data = {
    'Sales': [200, 250, 300, 280, 320],
    'Profit': [20, 30, 45, 40, 50],
    'Discount': [5, 10, 15, 10, 5],
    'Quantity': [2, 3, 4, 3, 5]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Calculating correlation matrix
corr_matrix = df.corr()

# Create heatmap
fig = px.imshow(
    corr_matrix,
    text_auto=True,            # show correlation values on cells
    color_continuous_scale='RdBu_r',  # similar to 'coolwarm'
    title='Correlation Heat Map'
)

# Display the heatmap inline in VS Code notebook
fig.show()
```

# 3. Scatter plots

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 13, 17, 20],
    'Z': [5, 8, 6, 9, 12],
    'Category': ['A', 'B', 'A', 'B', 'A']  # optional for color
}

df = pd.DataFrame(data)

# Create 3D scatter plot
fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    color='Category',   # optional, removes if you don't want color
    size='Y',           # optional, size of marker proportional to Y
    title='3D Scatter Plot Example'
)

# Display the plot inline in VS Code notebook
fig.show()
```
