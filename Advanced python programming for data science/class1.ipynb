```python
def my_generator(n):
    for i in range(n):
        yield i * i
gen = my_generator(3)
for value in gen:
    print(value)    
```

```python
def func(n):
    for i in range(n):
        return i
a = func(3)
type(a)
```

```python
def func(n):
    a = []
    for i in range(n):
        a.append(i)
    return a
a = func(3)
print(type(a))
for i in a:
    print(i)
```

```python
class MyClass:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def hello(self):
        print("hello")
M1 = MyClass("Ram", 22)
print(M1.name)
M2 = MyClass("Shyam", 23)
print(M2.name)
```
