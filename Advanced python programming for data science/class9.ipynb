Write a Python program that schedules a job to run repeatedly and prints "Hello World" every 5 seconds. You may use either the schedule module or Python’s built-in time module.

```python
import schedule
import time

# Function (job) to run
def my_job():
    print("Hello World")

# Schedule the job
# Runs every 5 seconds (you can change this)
schedule.every(5).seconds.do(my_job)

# Keep the program running
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
import schedule
import time
import pandas as pd
import sqlite3
import os

# -------------------------
# Folder where CSV & DB are stored
# -------------------------
data_folder = r'D:\SEMESTER 5 pratical\Advanced python programming for data science\data'
os.makedirs(data_folder, exist_ok=True)

# CSV file you downloaded
file_name = 'winemag-data-130k-v2.csv'
file_path = os.path.join(data_folder, file_name)

# -------------------------
# Function to clean CSV and save to SQLite
# -------------------------
def clean_and_store():
    print("Job started...")

    # 1️⃣ Read CSV
    try:
        df = pd.read_csv(file_path)
        print(f"CSV loaded from {file_path}")
    except Exception as e:
        print("Error reading CSV:", e)
        return

    # 2️⃣ Clean dataset
    try:
        df_clean = df.dropna()              # Remove missing values
        df_clean = df_clean.drop_duplicates()  # Remove duplicates
        print("Dataset cleaned.")
    except Exception as e:
        print("Error cleaning dataset:", e)
        return

    # 3️⃣ Store in SQLite
    try:
        db_path = os.path.join(data_folder, 'kaggle_data.db')
        conn = sqlite3.connect(db_path)
        df_clean.to_sql('wine_reviews', conn, if_exists='replace', index=False)
        conn.close()
        print(f"Cleaned data stored in SQLite at {db_path}")
    except Exception as e:
        print("Error saving to SQLite:", e)
        return

    print("Job finished.\n")

# -------------------------
# Schedule job every 5 seconds
# -------------------------
schedule.every(5).seconds.do(clean_and_store)

print("Scheduler started. Running every 5 seconds...")

while True:
    schedule.run_pending()
    time.sleep(1)
```
