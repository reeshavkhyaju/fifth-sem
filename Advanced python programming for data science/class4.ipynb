Write python program that reads a large text file and divied the data into chunks and process it using a generator.

```python
def read_file_in_chunks(file_path, chunk_size=1024):
    """
    Generator function that reads a file in chunks.
    
    :param file_path: Path to the text file
    :param chunk_size: Number of bytes per chunk
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        while True:
            chunk = file.read(chunk_size)
            if not chunk:
                break
            yield chunk


def process_chunk(chunk):
    """
    Example processing function.
    Here we count the number of words in a chunk.
    """
    return len(chunk.split())


# -------- Main Program --------
file_path = "D:\\SEMESTER 5 pratical\\Advanced python programming for data science\\large_text_file.txt"

total_words = 0

for chunk in read_file_in_chunks(file_path, chunk_size=4096):
    words_in_chunk = process_chunk(chunk)
    total_words += words_in_chunk

print("Total words in file:", total_words)
```

```python
def read_file_in_chunks(file_path, chunk_size=1):
    """
    Generator that reads a file in chunks of lines.
    """
    with open(file_path, "r", encoding="utf-8") as file:
        chunk = []
        for line in file:
            chunk.append(line.strip())
            if len(chunk) == chunk_size:
                yield chunk      # lazy yield
                chunk = []
        if chunk:
            yield chunk


def process_chunks(file_path, keyword):
    """
    Processes each chunk lazily.
    """
    total_count = 0

    for chunk in read_file_in_chunks(file_path):
        # process chunk
        count = sum(1 for line in chunk if keyword in line)
        total_count += count

    return total_count


# 🔹 Usage
if __name__ == "__main__":
    file_path = "D:\\SEMESTER 5 pratical\\Advanced python programming for data science\\large_text_file.txt"
    keyword = "ERROR"

    result = process_chunks(file_path, keyword)
    print(f"Lines containing '{keyword}': {result}")
```
