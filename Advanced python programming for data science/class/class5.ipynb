## Merge

```python
import pandas as pd

df1 = pd.read_excel('d1.xlsx')
df2 = pd.read_excel('d2.xlsx')

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
```

```python
df3 = df1.merge(df2)
print("\nMerged DataFrame:")
print(df3)
```

```python
df3_name = df1.merge(df2 , on='Name')
print("\nMerged DataFrame:")
print(df3_name)
```

```python
df3_inner = df1.merge(df2 ,how='left')
print("\nMerged DataFrame:")
print(df3_inner)
```

```python
df3_inner = df1.merge(df2 ,how='right')
print("\nMerged DataFrame:")
print(df3_inner)
```

```python
df3_inner = df1.merge(df2 ,how='outer')
print("\nMerged DataFrame:")
print(df3_inner)
```

```python
df3_inner = df1.merge(df2 ,how='inner')
print("\nMerged DataFrame:")
print(df3_inner)
```

## Reshaping

```python
import pandas as pd

# -----------------------------
# Step 1: Create sample dataset
# -----------------------------
data = {
    'Student': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 90, 78],
    'Science': [92, 88, 95],
    'English': [88, 79, 85]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

# -----------------------------
# Step 2: Melt the DataFrame
# -----------------------------
melted_df = pd.melt(df, id_vars=['Student'], var_name='Subject', value_name='Score')

print("\nMelted DataFrame:\n")
print(melted_df)
```

```python
import pandas as pd

# -------------------------------
# Create WIDE dataset
# -------------------------------
data = {
    "subject_id": [1, 2, 3, 4, 5],
    "group": ["Control", "Control", "Treatment", "Treatment", "Treatment"],
    "score_t1": [65, 60, 70, 72, 68],
    "score_t2": [68, 62, 75, 76, 71],
    "score_t3": [70, 64, 78, 79, 74],
    "score_t4": [72, 66, 80, 83, 77]
}

df_wide = pd.DataFrame(data)
print("WIDE DATASET")
print(df_wide)

# -------------------------------
# Melt to LONG format
# -------------------------------
df_long = pd.melt(
    df_wide,
    id_vars=["subject_id", "group"],
    value_vars=["score_t1", "score_t2", "score_t3", "score_t4"],
    var_name="time",
    value_name="score"
)

print("\nMELTED (LONG) DATASET")
print(df_long)
```

```python
import pandas as pd

# -------------------------------
# Create WIDE dataset
# -------------------------------
data = {
    "subject_id": [1, 2, 3, 4, 5],
    "group": ["Control", "Control", "Treatment", "Treatment", "Treatment"],
    "score_t1": [65, 60, 70, 72, 68],
    "score_t2": [68, 62, 75, 76, 71],
    "score_t3": [70, 64, 78, 79, 74],
    "score_t4": [72, 66, 80, 83, 77]
}

df_wide = pd.DataFrame(data)
print("WIDE DATASET")
print(df_wide)

# -------------------------------
# WIDE → LONG (MELT)
# -------------------------------
df_long = pd.melt(
    df_wide,
    id_vars=["subject_id", "group"],
    value_vars=["score_t1", "score_t2", "score_t3", "score_t4"],
    var_name="time",
    value_name="score"
)

print("\nLONG (MELTED) DATASET")
print(df_long)

# -------------------------------
# LONG → WIDE (PIVOT)
# -------------------------------
df_pivot = df_long.pivot(
    index=["subject_id", "group"],
    columns="time",
    values="score"
).reset_index()

print("\nPIVOTED (WIDE) DATASET")
print(df_pivot)
```
