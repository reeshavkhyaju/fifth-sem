```python

df1 = pd.read_csv('csv1.csv')
df2 = pd.read_csv('csv2.csv')
df3 =df1.merge(df2, how='left')
print(df3)
```

```python

df1 = pd.read_csv('csv1.csv')
df2 = pd.read_csv('csv2.csv')
df3 =df1.merge(df2, how='right')
print(df3)
```

```python

df1 = pd.read_csv('csv1.csv')
df2 = pd.read_csv('csv2.csv')
df3 =df1.merge(df2, how='outer')
print(df3)
```

```python

df1 = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [11,12,13,14]
})
df2 = pd.DataFrame({
    "C": [5, 6],
    "D": [16,17,]
})
df3 = df1.join(df2)
print(df3)
```

```python

df1 = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [11,12,13,14]
}, index=['a','b','c','d'])
df2 = pd.DataFrame({
    "C": [5, 6],
    "D": [16,17,]
}, index=['a','b'])
df3 = df1.join(df2, how = 'left' )
print(df3)
```

```python

df1 = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [11,12,13,14]
})
df2 = pd.DataFrame({
    "C": [5, 6],
    "D": [16,17,]
})
df3 = df1.join(df2, how = 'right')
print(df3)
```
