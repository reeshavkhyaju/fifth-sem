1. Line Plot

```python
import matplotlib.pyplot as plt

# 1. Define the data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 180, 170, 200, 220]

# 2. Create the line plot with circular markers ('o')
plt.plot(months, sales, marker='o')

# 3. Add a title and axis labels
plt.title('Monthly Sales of the Company')
plt.xlabel('Months')
plt.ylabel('Sales')

# 4. Enable the background grid for better readability
plt.grid(True)

# 5. Display the resulting graph
plt.show()
```

2. Bar chart

```python
import matplotlib.pyplot as plt

# Given data
departments = ['CSE', 'ECE', 'ME', 'CE']
students = [120, 90, 60, 50]

# Creating the bar chart
plt.bar(departments, students)

# Adding title and labels
plt.title('Number of Students in Different Departments')
plt.xlabel('Departments')
plt.ylabel('Number of Students')

# Display the chart
plt.show()
```

3. Histogram 

```python
import matplotlib.pyplot as plt

# Given marks
marks = [45, 55, 60, 70, 75, 80, 85, 90, 65, 68, 72, 78]

# Plotting the histogram
plt.hist(marks, bins=5)

# Adding title and labels
plt.title('Histogram of Students Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')

# Display the histogram
plt.show()
```

4. Scatter Plot 

```python
import matplotlib.pyplot as plt

# Given data
hours_studied = [1, 2, 3, 4, 5, 6]
marks = [35, 40, 50, 65, 75, 85]

# Creating the scatter plot
plt.scatter(hours_studied, marks)

# Adding title and labels
plt.title('Relationship Between Hours Studied and Marks Obtained')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')

# Display the plot
plt.show()
```

5. Subplots

```python
import matplotlib.pyplot as plt

# Data for Subplot 1: Monthly Sales (Line Plot)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 150, 180, 170, 200, 220]

# Data for Subplot 2: Department-wise Students (Bar Chart)
departments = ['CSE', 'ECE', 'ME', 'CE']
students = [120, 90, 60, 50]

# Creating subplots
plt.figure(figsize=(10, 4))

# Subplot 1: Line Plot
plt.subplot(1, 2, 1)
plt.plot(months, sales, marker='o')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)

# Subplot 2: Bar Chart
plt.subplot(1, 2, 2)
plt.bar(departments, students)
plt.title('Department-wise Student Strength')
plt.xlabel('Departments')
plt.ylabel('Number of Students')

# Adjust layout
plt.tight_layout()

# Display the figure
plt.show()
```

q. Why matplotlib.pyplot is imported?
- We use matplotlib.pyplot to:
1. set the title
2. set X-axis and Y-axis labels
3. show the plot using plt.show()

1. Box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
marks = {
    'Mathematics': [65, 70, 75, 80, 85, 90],
    'Physics': [60, 68, 72, 78, 82, 88],
    'Chemistry': [62, 67, 74, 79, 84, 89]
}

# Creating the box plot
sns.boxplot(data=marks)

# Adding title and labels
plt.title('Distribution of Marks in Different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')

# Display the plot
plt.show()
```

2. Pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Creating the dataset
data = {
    'Height': [150, 155, 160, 165, 170],
    'Weight': [50, 55, 60, 65, 70],
    'Age': [18, 19, 20, 21, 22]
}

df = pd.DataFrame(data)

# Creating the pair plot
sns.pairplot(df)

# Adding title
plt.suptitle('Pair Plot of Height, Weight, and Age', y=1.02)

# Display the plot
plt.show()
```

3. Heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    'Sales': [200, 250, 300, 280, 320],
    'Profit': [20, 30, 45, 40, 50],
    'Discount': [5, 10, 15, 10, 5],
    'Quantity': [2, 3, 4, 3, 5]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Calculating correlation matrix
corr_matrix = df.corr()

# Plotting heat map
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Adding title
plt.title('Correlation Heat Map')

# Display the plot
plt.show()
```
