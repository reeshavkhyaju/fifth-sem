```python
import plotly.express as px

df = px.data.gapminder().query("country=='Nepal'")
fig = px.line(df, x="year", y="lifeExp", title="Life Expectancy in Nepal")
fig.show()
```

```python
df = px.data.iris()
fig = px.bar(df, x="species", y="sepal_length", title="Sepal Length by Species")
fig.show()
```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    title="Sepal Width vs Length"
)
fig.show()
```

```python
fig = px.histogram(df, x="sepal_length", title="Distribution of Sepal Length")
fig.show()
```

```python
fig = px.box(df, x="species", y="sepal_length", title="Sepal Length Spread")
fig.show()
```

```python
df2 = df["species"].value_counts().reset_index()
df2.columns = ["species", "count"]

fig = px.pie(df2, names="species", values="count", title="Species Distribution")
fig.show()
```

```python
fig = px.area(df, x=df.index, y="sepal_length", title="Area Chart Example")
fig.show()
```

```python
corr = df.corr(numeric_only=True)
fig = px.imshow(corr, text_auto=True, title="Correlation Heatmap")
fig.show()
```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    size="petal_length",
    color="species",
    title="Bubble Chart"
)
fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="sepal_length",
    y="sepal_width",
    z="petal_length",
    color="species",
    title="3D Scatter Plot"
)
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=[1,2,3], y=[2,4,6], mode="lines", name="Line"))
fig.add_trace(go.Bar(x=[1,2,3], y=[2,4,6], name="Bar"))

fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            buttons=[
                dict(label="Line",
                     method="update",
                     args=[{"visible": [True, False]}]),
                dict(label="Bar",
                     method="update",
                     args=[{"visible": [False, True]}])
            ]
        )
    ]
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# -----------------------------
# Generate synthetic data
# -----------------------------
np.random.seed(42)

num_particles = 30
num_frames = 60
t = np.linspace(0, 2 * np.pi, num_frames)

# Particle trajectories
x = np.array([np.cos(t + phase) for phase in np.linspace(0, 2, num_particles)])
y = np.array([np.sin(t + phase) for phase in np.linspace(0, 2, num_particles)])

# -----------------------------
# Initial figure
# -----------------------------
fig = go.Figure(
    data=[
        go.Scatter(
            x=x[:, 0],
            y=y[:, 0],
            mode="markers",
            marker=dict(
                size=10,
                color=np.arange(num_particles),
                colorscale="Viridis",
                showscale=True
            ),
            hovertemplate="Particle %{marker.color}<br>x=%{x:.2f}<br>y=%{y:.2f}"
        )
    ],
    layout=go.Layout(
        title="Plotly Native Frame-Based Animation",
        xaxis=dict(range=[-1.5, 1.5], autorange=False),
        yaxis=dict(range=[-1.5, 1.5], autorange=False),
        updatemenus=[
            dict(
                type="buttons",
                showactive=False,
                buttons=[
                    dict(
                        label="▶ Play",
                        method="animate",
                        args=[
                            None,
                            dict(
                                frame=dict(duration=60, redraw=True),
                                transition=dict(duration=30),
                                fromcurrent=True,
                                mode="immediate",
                            ),
                        ],
                    ),
                    dict(
                        label="⏸ Pause",
                        method="animate",
                        args=[
                            [None],
                            dict(frame=dict(duration=0), mode="immediate"),
                        ],
                    ),
                ],
            )
        ],
        sliders=[
            dict(
                active=0,
                steps=[],
                x=0.1,
                y=0,
                len=0.9,
            )
        ],
    ),
)

# -----------------------------
# Create animation frames
# -----------------------------
frames = []
slider_steps = []

for i in range(num_frames):
    frame = go.Frame(
        data=[
            go.Scatter(
                x=x[:, i],
                y=y[:, i],
            )
        ],
        name=str(i),
    )
    frames.append(frame)

    slider_steps.append(
        dict(
            method="animate",
            args=[
                [str(i)],
                dict(
                    mode="immediate",
                    frame=dict(duration=60, redraw=True),
                    transition=dict(duration=30),
                ),
            ],
            label=f"{i}",
        )
    )

fig.frames = frames
fig.layout.sliders[0]["steps"] = slider_steps

# -----------------------------
# Show animation
# -----------------------------
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Generate large synthetic data (100,000 points)
n = 100000
time = np.linspace(0, 10, n)
value = np.sin(time) + np.random.normal(0, 0.1, n)

# Using Scattergl for high-performance WebGL rendering
fig = go.Figure(data=go.Scattergl(
    x=time,
    y=value,
    mode="markers",
    marker=dict(
        size=3,
        color=value,
        colorscale="Viridis",
        opacity=0.6
    )
))

fig.update_layout(
    title=f"High Performance WebGL Plot ({n:,} points)",
    xaxis_title="Time",
    yaxis_title="Value"
)

fig.show()
```
